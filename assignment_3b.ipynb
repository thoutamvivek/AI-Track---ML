-->Find the global minimum point and value for the function f(x) = 3x^2+5e^-y+10?

```python
x_o = 2
y_o =3
eta = 0.9
eps = 0.000001
del_x = 1
del_y=1
max_iterations = 10000
iterations = 0
def derivation(x,y):
    x_derivation = 6*(x)
    y_derivation = 5
    return x_derivation,y_derivation
while max(abs(del_x),abs(del_y)) > eps and iterations < max_iterations:
    prev_x = x_o
    prev_y = y_o
    del_x,del_y = derivation(prev_x,prev_y)
    del_x=-eta*del_x
    del_y=-eta*del_y
    x_o = x_o + del_x
    y_o = y_o + del_y
    iterations = iterations+1
    print(iterations,"Iterations","\n x value is",x_o,"\n y value is",y_o)
print("The local minimum occurs at",x_o,y_o)
```
